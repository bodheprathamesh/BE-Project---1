```python
df = df.drop(['Name','Peak Season','Before Season months', 'Months left for End of Season','Gender'],axis = 1)
```

```python
df = df.drop(['TransactionCount','Unnamed: 0'],axis = 1)
```

```python
x = df.iloc[:,0:16]
y = df.iloc[:,16:17]
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)    
y_train_scaled = scaler.fit_transform(y_train)
y_test_scaled = scaler.transform(y_test)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, LSTM
from keras.optimizers import Adam
```

```python
classifier = Sequential()
```

```python
classifier.add(LSTM(units=128, activation='relu', input_shape=(16, 1)))
# classifier.add(Dense(128, activation = 'relu', input_dim=X_train_scaled.shape[1]))
```

```python
classifier.add(Dense(64, activation = 'relu'))
```

```python
classifier.add(Dense(1,activation='linear'))
```

```python
classifier.compile(optimizer='adam', loss = 'mean_squared_error', metrics=['mean_squared_error'])
```

```python
from tensorflow.keras.utils import plot_model
plot_model(classifier, to_file='neural_network.png', show_shapes=True, show_layer_names=True)
```

```python
import matplotlib.pyplot as plt
def visualize_model(model):
    fig, ax = plt.subplots(figsize=(12, 8))
    plot_model(model, to_file='model_plot.png', show_shapes=True, show_layer_names=True, rankdir='TB', expand_nested=True)
    img = plt.imread('model_plot.png')
    ax.imshow(img)
    ax.axis('off')
    plt.show()

# Visualize the model
visualize_model(classifier)
```

```python
classifier.fit(X_train_scaled, y_train_scaled, batch_size = 256, epochs = 64, validation_split = 0.2, verbose = 2)
```

```python
y_pred = classifier.predict(X_test_scaled)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
print(mean_squared_error(y_test_scaled,y_pred))
print(mean_absolute_error(y_test_scaled,y_pred))
```

```python
y_test_flatten = y_test_scaled.flatten()
y_pred_flatten = y_pred.flatten()
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(x=y_test_flatten, y=y_pred_flatten)
plt.title('True Values vs Predicted Values')
plt.xlabel('True Values')
plt.ylabel('Predicted Values')
plt.show()
```
